```python
import dask.dataframe as dd
import csv
import pandas as pd
```

```python
dtype = {
    'candidate_id': 'object',
    'device': 'object',
    'favorite_count': 'object',
    'id': 'object',
    'latitude': 'object',
    'longitude': 'object',
    'polarity': 'object',
    'retweet_count': 'object',
    'retweeted_status_id': 'object',
    'subjectivity': 'object',
    'tw_user_id': 'object',
    'tweet_id': 'object',
    'state': 'object' 
}

# Define chunk size
chunk_size = 10000

# Create an empty DataFrame to store the results
all_tweets = pd.DataFrame()

# Read the CSV file in chunks
file_path = "C:/Users/kesha/Downloads/20zn 16 US Presidential Elections Tweets CSV/2016 US Presidential Elections Tweets CSV/2016_US_election_tweets.csv"
for chunk in pd.read_csv(file_path, dtype=dtype, usecols=['tweet_text'], chunksize=chunk_size):
    # Process each chunk as needed
    all_tweets = pd.concat([all_tweets, chunk], ignore_index=True)
```

```python
all_tweets
```

```python
trump_keyword = '@realdonaldtrump'
clinton_keyword = '@hillaryclinton'

# Filter tweets based on keywords
filtered_tweets = all_tweets[all_tweets['tweet_text'].str.contains(trump_keyword, case=False) | all_tweets['tweet_text'].str.contains(clinton_keyword, case=False)]


filtered_tweets['Candidate'] = ''
filtered_tweets.loc[filtered_tweets['tweet_text'].str.contains(trump_keyword, case=False), 'Candidate'] = 'Trump'
filtered_tweets.loc[filtered_tweets['tweet_text'].str.contains(clinton_keyword, case=False), 'Candidate'] = 'Clinton'
```

```python
filtered_tweets
```

```python
import pandas as pd

# Sample 1 million tweets for each candidate
trump_tweets = filtered_tweets[filtered_tweets['Candidate'] == 'Trump'].sample(n=1000000, random_state=42)
clinton_tweets = filtered_tweets[filtered_tweets['Candidate'] == 'Clinton'].sample(n=1000000, random_state=42)

# Concatenate the sampled DataFrames
df = pd.concat([trump_tweets, clinton_tweets], ignore_index=True)

# Shuffle the rows in the new DataFrame
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
df
```

```python
df.to_csv("2016_Senti.csv")
```
